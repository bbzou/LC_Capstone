```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
safe_df=pd.read_csv('safe_df.csv',index_col=None)
safe_df.drop(['Unnamed: 0'],axis=1,inplace=True)
safe_df=safe_df.sort_values(by='return_rate',ascending=False).iloc[0:50,:]
safe_df.reset_index(drop=True,inplace=True)
safe_df.drop(['application_type'],axis=1,inplace=True)
```
